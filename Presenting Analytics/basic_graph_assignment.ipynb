```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
from matplotlib.pyplot import figure
import numpy as np





df_Opiods = pd.read_csv('C:/Users/thoma/OneDrive/Documents/CIBA/CTBA/CTBA/M2/OpidsVA.csv') #bring in data




ax1 = df_Opiods[['MedianHouseholdIncome']].plot(kind='bar',width=.1) #creates bar chart
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left") #puts legend in upper left
ax2 = df_Opiods['FPOO-Rate'].plot(secondary_y=True, color = 'red') #creates line chart

red_patch = mpatches.Patch(color='red',) #makes line chart red
plt.legend(handles=[red_patch]) #adds red axis

plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")


plt.suptitle('Median Household Income Compared to Number of Opiod Deaths for Counties in VA', fontsize=12) #names titles at axis
ax1.set_xlabel('FIPSCODE', fontsize=10)
ax2.set_ylabel('Opiod Deaths by County', fontsize=10)
ax1.set_ylabel('Median Income', fontsize=10)



fig = plt.gcf()
fig.set_size_inches(18.5, 10.5) #sizes axis





ax1 = plt.show()

#I combined a bar chart containing Median Income Data and a line chart that shows # of opiod deaths; As the median income drops, the number of opiod deaths increase
   

```

```python
df_pres = pd.read_csv('C:/Users/thoma/OneDrive/Documents/CIBA/CTBA/CTBA/M2/presidents.csv') #bring in data

import pandas
from collections import Counter
a = (df_pres['State']) #set state column equal to a
letter_counts = Counter(a) #counts the number of iterations of the state
df = pandas.DataFrame.from_dict(letter_counts, orient='index') 
ax = df.plot(kind='bar',legend = 0) #creates bar chart

ax.set_ylabel("number of presidents ") #ylabel
ax.set_xlabel("States") #xlabel
plt.suptitle('States Where US Presidents Were Born', fontsize=12) #names titles at axis

df_pres #creates dataframe

#Virginia has the greatest number of presidents who hail from their state, per the bar graph I created

```

```python
df_NSA = pd.read_csv('C:/Users/thoma/OneDrive/Documents/CIBA/CTBA/CTBA/M2/TOTALNSA.csv') 
Year = df_NSA['DATE'].str[:4] #strip the year from the date column

df_NSA['Year'] = Year #add column to df

df_NSA['DATE'] = pd.to_datetime(df_NSA.DATE)
Yearly_AVG= df_NSA.groupby(df_NSA.DATE.dt.year)['TOTALNSA'].transform('mean') #get yearly average
df_NSA['Yearly_AVG'] = Yearly_AVG #add column to df
df_drop=df_NSA.drop_duplicates(subset='Year', keep="last") #drop duplicates



ax = df_drop.set_index('Year')['Yearly_AVG'].plot()

ax.set_ylabel("Average Number of Cars Sold ") #ylabel
ax.set_xlabel("Year") #xlabel
plt.suptitle('Average auto sales in the US from 1976-2018', fontsize=12) #names titles at axis

df_drop

#Using a line chart over the yearly average, we find that autosales trend down based off recessions, for example 2008 and early 1980's
```
